```python
def set_rank(n, m, r):
    if r > min(n, m):
        raise ValueError("Rank r cannot be greater than number of columns m")
    A = np.random.randn(n, m)
    B = np.random.randn(m, r) @ np.random.randn(r, m)
    C = A @ B

    return C


F = np.random.randn(9, 18)
R = set_rank(9, 18, 7)
l = np.random.randint(1, 10, 1)

print("Rank of full-rank matrix:", np.linalg.matrix_rank(F))
print("Rank of reduced-rank matrix:", np.linalg.matrix_rank(R))
print("Scalar:", l)

print("\nRank of scalar multiplied full-rank matrix:", np.linalg.matrix_rank(l * F))
print("Rank of scalar multiplied reduced-rank matrix:", np.linalg.matrix_rank(l * R))

if np.linalg.matrix_rank(l * R) == l * np.linalg.matrix_rank(R):
    print(
        f"\nrank(l*r)({np.linalg.matrix_rank(l*R)}) = l*rank(R)({l*np.linalg.matrix_rank(R)})"
    )
else:
    print(
        f"\nrank(l*r)({np.linalg.matrix_rank(l*R)}) != l*rank(R)({l*np.linalg.matrix_rank(R)})"
    )
```

## Code Challenge 2: Summed and multiplied matrix rank.

Rules: 

- $rank(AB) \ge min(rank(A), rank(B))$
- $rank(A+B) \ge rank(A)+rank(b)$

Challenge:

- Generate two matrices (A and B), 2 x 5
- Compute $A^TA$ and $B^TB$, and find rank.
- Find $A^TA$ x $B^TB$ and $A^TA$ + $B^TB$, and compute the rank.



```python
import numpy as np

A = np.random.randn(2, 5)
print(f"Original matrix A: \n{A}")

B = np.random.randn(2, 5)
print(f"Original matrix B: \n{B}")

rankA = np.linalg.matrix_rank(A)
print(f"Rank of A: {rankA}")

rankB = np.linalg.matrix_rank(B)
print(f"Rank of B: {rankB}")

AtA = A.T @ A
print(f"Rank of A^T A: {np.linalg.matrix_rank(AtA)}")

BtB = B.T @ B
print(f"Rank of B^T B: {np.linalg.matrix_rank(BtB)}")

print(f"Rank of A^T A @ B^T B: {np.linalg.matrix_rank(AtA @ BtB)}")
print(f"Rank of A^T A + B^T B: {np.linalg.matrix_rank(AtA + BtB)}")
```

## Code Challenge 3: Is a vector within the span of a set?
Is this vector:
$v = \begin{bmatrix} 1 \\ 2 \\ 3 \\ 4 \end{bmatrix}$ 

In these sets:

$S = \begin{bmatrix} 4 & 0 \\ 3 & 4 \\ 6 & 0 \\ 2 & 1  \end{bmatrix}$ 
$T = \begin{bmatrix} 1 & 0 \\ 2 & 0 \\ 2 & 1 \\ 2 & 2  \end{bmatrix}$ 



```python
v = np.array([1, 2, 3, 4]).T
print(f"Original vector v: {v}")

S = np.vstack([[4, 3, 6, 2], [0, 4, 0, 1]]).T
print(f"Matrix S: \n{S}")

T = np.vstack([[1, 2, 2, 2], [0, 0, 1, 2]]).T
print(f"Matrix T: \n{T}")

print("Span of S and v:")
print(np.linalg.matrix_rank(np.column_stack((S, v))))


print("Span of T and v:")
print(np.linalg.matrix_rank(np.column_stack((T, v))))
```
